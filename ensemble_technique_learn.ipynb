````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#sklearn
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.tree import plot_tree, DecisionTreeRegressor
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import Lasso
from sklearn.ensemble import VotingRegressor
from sklearn.svm import SVR
from sklearn.ensemble import AdaBoostRegressor,GradientBoostingRegressor, StackingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
````

````python
mdl1=LinearRegression()
mdl2=RandomForestRegressor(n_estimators=100)
mdl3=SVR()
````

````python
_data=load_boston()
x=_data['data']
y=_data['target']
````

````python
fig,_ax=plt.subplots(1,1,figsize=(20,7))
sns.heatmap(pd.DataFrame(x).corr().abs(),annot=True,ax=_ax)
````

````python
x_train,x_test,y_train,y_test=train_test_split(x[:,upper.columns],y,test_size=0.25,random_state=14)
````

**bagging- random forest**

````python
_random_forest=RandomForestRegressor(max_depth=100)
_random_forest.fit(x_train,y_train)
r2_score(y_test,_random_forest.predict(x_test))
````

**voting**

````python
voting_reg=VotingRegressor([
                            ('l1',mdl1),
                            ('l2',mdl2),
                            ('l3',mdl3)
])
voting_reg.fit(x_train,y_train)
r2_score(y_test,voting_reg.predict(x_test))
````

````python
ada_boost=AdaBoostRegressor()
ada_boost.fit(x_train,y_train)
r2_score(y_test,ada_boost.predict(x_test))
````

**gradient boost**

````python
gradient_boost=GradientBoostingRegressor()
gradient_boost.fit(x_train,y_train)
r2_score(y_test,gradient_boost.predict(x_test))
````

**stacking**

````python
stacking_reg=StackingRegressor(estimators=[
                                           ('l1',mdl1),
                                           ('l2',mdl3)
],final_estimator=mdl2)
stacking_reg.fit(x_train,y_train)
r2_score(y_test,stacking_reg.predict(x_test))
````

````python
pipes=Pipeline([
                ('scaling2',StandardScaler()),
                ('mdl','passthrough')
])
_estimators=[
                                  ('model1',LinearRegression()),
                                  ('model2',SVR()),
                                  ('model3',DecisionTreeRegressor())
               ]
_final_estimator=[
              LinearRegression(),
                SVR()
]
parameters=[
             {
              'mdl':[StackingRegressor(estimators=_estimators)],
              'mdl__model3__max_depth':[0.4,0.6,0.8,1],
              'mdl__final_estimator':_final_estimator
            },
            {
                'mdl':[voting_reg]
            }
]

grid_search=GridSearchCV(pipes,parameters,n_jobs=-1,cv=5,scoring='r2')
grid_search.fit(x_train,y_train)
````
